```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("IRIS2.csv")
df
```

```python
df.isnull().sum()
```

```python
ndf=df
# ndf['sepal_length']=ndf['sepal_length'].fillna(ndf['sepal_length'].mean())
ndf=ndf.fillna(ndf.mean(numeric_only=True))

ndf
```

```python
df.boxplot(column="sepal_length")
plt.show()
```

```python
col=['sepal_length','petal_length']
df.boxplot(col)
plt.show()
```

```python
#removal of outlier
Q1=df['sepal_length'].quantile(0.25)
Q3=df['sepal_length'].quantile(0.75)
IQR=Q3-Q1
print("IQR:",IQR)
print("shape before ",df.shape)

df=df[~((df['sepal_length']<(Q1-IQR*1.5))|(df['sepal_length']>(Q3+1.5*IQR)))]
print("shape after ",df.shape)

df.boxplot(['sepal_length','petal_length'])
plt.show()
```

```python
x=df.sepal_length
y=df.petal_length
plt.plot(x,y)
plt.show()
```
